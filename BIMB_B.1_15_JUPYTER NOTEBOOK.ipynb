```python
df4 = pd.read_csv('energy4_csv.csv')
new_df4 = df4.groupby('Country').agg({'Population': 'sum', 'Energy_intensity_per_capita': 'sum'}).reset_index()
new_df4.plot(kind='bar', x='Country', y=['Population', 'Energy_intensity_per_capita'], figsize=(12,8)) 
#low energy consumption
```

```python
df4 = pd.read_csv('energy4_csv.csv')
new_df4 = df4.groupby('Country').agg({'Energy_consumption': 'sum', 'CO2_emission': 'sum'}).reset_index()
new_df4.plot(kind='bar', x='Country', y=['Energy_consumption', 'CO2_emission'], figsize=(12,8)) 
#disproportionate co2 emissions compared to consumption (very polluting to extract petroleum and other liquids)
```

```python
import pandas as pd

df8 = pd.read_csv('energy8_csv.csv') 
```

```python
all_energy = df8[df8['Energy_type'] == 'all_energy_types']
petroleum = df8[df8['Energy_type'] == 'petroleum_n_other_liquids']
merged = pd.merge(all_energy, petroleum, on='Country', suffixes=('_all_energy', '_petroleum'))
ax = merged.plot(kind='bar', x='Country', y=['Energy_production_all_energy', 'Energy_production_petroleum'], rot=45)
ax.set_xlabel('Country')
ax.set_ylabel('Energy production (kWh)') #allows us to put the names sideway to have a cleaner look at the graph
```
